```
# File location and type

# file_location = "/FileStore/tables/tripdata_full.csv"
file_location = "/FileStore/tables/tripdata.csv"
file_type = "csv"

# CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(df)

```

```
# Fixing the column names

# Chain withColumnRenamed() methods to rename columns
df = df.withColumnRenamed("start_time", "start_time_h") \
    .withColumnRenamed("_c1", "start_time_m") \
    .withColumnRenamed("_c2", "start_time_s") \
    .withColumnRenamed("_c3", "start_time_am_pm") \
    .withColumnRenamed("end_time", "end_time_h") \
    .withColumnRenamed("_c5", "end_time_m") \
    .withColumnRenamed("_c6", "end_time_s") \
    .withColumnRenamed("_c7", "end_time_am_pm") \
    .withColumnRenamed("pyment", "payment")

display(df)
```

```
# Schema looks good
df.printSchema()
```

```
from pyspark.sql.functions import count, when, isnull

null_counts = df.select([count(when(isnull(c), c)).alias(c) for c in df.columns]).toPandas().transpose()
#df.isnull().sum

# WE NEED TO BRAINSTORM ABT HOW TO FILL THE NULL ROWS!
print(null_counts)
```

```
# checking the mode of the gender
from pyspark.sql.functions import mode
df.select(mode('member_gender')).collect()
```

```
# assigning the new values
df = df.fillna(value=1983, subset=['member_birth_year'])
df = df.fillna(value='Male', subset=['member_gender'])
# checking if the assignment is successfull
null_counts = df.select([count(when(isnull(c), c)).alias(c) for c in df.columns]).toPandas().transpose()
print(null_counts)
# SUCSESS
```

```
# Creating a timestamp column of start and end time
from pyspark.sql.functions import col, to_timestamp, concat_ws, format_string, date_format

df = df.withColumn("start_time", to_timestamp(concat_ws(":", format_string("%02d",col("start_time_h")),
                                                        format_string("%02d", col("start_time_m")), format_string("%02d", col("start_time_s"))), "HH:mm:ss")) \
       .withColumn("start_time", date_format("start_time", "HH:mm:ss")) \
       .withColumn("end_time", to_timestamp(concat_ws(":", format_string("%02d",col("end_time_h")), format_string("%02d",col("end_time_m")), format_string("%02d", col("end_time_s"))), "HH:mm:ss")) \
       .withColumn("end_time", date_format("end_time", "HH:mm:ss")) 

display(df)
```

```
cond = col("start_time") > col("end_time")

# Use an auxiliary variable to swap start_date and end_date if necessary
df = df.withColumn("aux_start_time", when(cond, col("end_time")).otherwise(col("start_time"))) \
       .withColumn("aux_end_time", when(cond, col("start_time")).otherwise(col("end_time"))) \
       .withColumn("start_time", col("aux_start_time")).withColumn("end_time", col("aux_end_time")) # Replace start_date and end_date with the corrected values

# Drop the auxiliary columns
df = df.drop("aux_start_time", "aux_end_time")


display(df)
```

```
df = df.dropDuplicates()
```

```
#No duplicates in the data which is nice

df.toPandas().info()
```

```
# Dropping columns that are redundant

df = df.drop("start_time_h", "start_time_m", "start_time_s", "start_time_am_pm", "end_time_h", "end_time_m", "end_time_s", "end_time_am_pm")

# Save the cleaned data

df.write.format("csv").mode("overwrite").option("header", "true").save("/FileStore/tables/clean_tripdata.csv")
```
